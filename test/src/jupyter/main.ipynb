```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Carlos\Desktop\projetos\JAVA_me_odeia\analisador_comparativo_av3\test\src\main\java\com\csv\DonQuixote_x_Quijote.csv", sep=",")
```

```python
df.head()
```

```python
df_serial = df[["Amostra" , "serial"]]
df_gpu = df[["Amostra", "gpu"]]
```

```python
df_serial
```

```python
# Configurando a figura
plt.figure(dpi=120, figsize=(10, 6))

# Plotando os dados
plt.plot(df_serial['Amostra'], df_serial['serial'], color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, label='Serial Data')

# Adicionando ttulo e rtulos
plt.title('Desempenho Serial vs Amostras', fontsize=16, fontweight='bold')
plt.xlabel('Amostras', fontsize=12)
plt.ylabel('Desempenho Serial em MS', fontsize=12)

# Adicionando grid para facilitar leitura
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Adicionando legenda
plt.legend(loc='best', fontsize=10)

# Ajustando os limites dos eixos, se necessrio
plt.xlim(df_serial['Amostra'].min() - .25, df_serial['Amostra'].max()+.25)
plt.ylim(df_serial['serial'].min() - 1, df_serial['serial'].max() + 1)

plt.xticks(ticks=[1, 2, 3], labels=["1", "2", "3"])

# Mostrando o grfico
plt.show()
```

```python
# Configurando a figura
plt.figure(dpi=120, figsize=(12, 8))

# Plotando os dados
plt.plot(df_gpu['Amostra'], df_gpu['gpu'], color='red', linestyle='-', linewidth=2, marker='o', markersize=6, label='gpu Data')

# Adicionando ttulo e rtulos
plt.title('Desempenho gpu vs Amostras', fontsize=16, fontweight='bold')
plt.xlabel('Amostras', fontsize=12)
plt.ylabel('Desempenho gpu em MS', fontsize=12)

# Adicionando grid para facilitar leitura
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Adicionando legenda
plt.legend(loc='best', fontsize=10)

# Ajustando os limites dos eixos, se necessrio
plt.xlim(df_gpu['Amostra'].min() - .25, df_gpu['Amostra'].max()+.25)
plt.ylim(df_gpu['gpu'].min() - 1, df_gpu['gpu'].max() + 1)

plt.xticks(ticks=[1, 2, 3], labels=["1", "2", "3"])

# Mostrando o grfico
plt.show()
```
